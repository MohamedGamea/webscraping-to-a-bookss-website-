```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
response
```

```python
soup = BeautifulSoup(response.text)
```

```python
articles = soup.find_all('article',class_="product_pod")
articles
```

check one article

```python
articles[0]
```

part one : find title

```python
art = articles[0]

h3 = art.find('h3')
a = h3.find('a')
title = a.attrs['title']

title
```

part 2 : find price

```python
price = art.find('p',class_='price_color').text
price = float(price[2:])

price
```

part 3 : find Rating

```python
rating = art.find('p',class_="star-rating")
rating =rating.attrs['class'][-1]

rating
```

```python
def convert(num):
    values = {
        'One':1,
        'Two':2,
        'Three':3,
        'Four':4,
        'Five':5
    }

    return values[num]
```

```python
for art in articles:

    h3 = art.find('h3')
    a = h3.find('a')
    title = a.attrs['title']

    price = art.find('p',class_='price_color').text
    price = float(price[2:])

    rating = art.find('p',class_="star-rating")
    rating =rating.attrs['class'][-1]

    num_rating = convert(rating)

    print(f"{title}, {price}, {num_rating}")
```

Export data to a csv file

```python
file_name = soup.find('div', class_="page-header action").text
file_name = file_name[1:-1] + '.csv'
file_name
```

```python
with open(file_name,'w' , encoding= 'utf8') as csv:
    csv.write('title\tprice\trating\n')

    for f in articles:
        h3 = f.find('h3')
        a = h3.find('a')
        title = a.attrs['title']

        price = f.find('p',class_='price_color').text
        price = float(price[2:])

        rating = f.find('p',class_="star-rating")
        rating =rating.attrs['class'][-1]

        num_rating = convert(rating)

        row = f"{title}\t{price}\t{num_rating}\n"

        csv.write(row)
```

```python
import pandas as pd
df = pd.read_csv(file_name, sep='\t')
df
```
